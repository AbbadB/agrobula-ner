```python
from lxml.html.clean import Cleaner
```

```python
def sanitize(dirty_html):
    cleaner = Cleaner(page_structure=True,
                  meta=True,
                  embedded=True,
                  links=True,
                  style=True,
                  processing_instructions=True,
                  inline_style=True,
                  scripts=True,
                  javascript=True,
                  comments=True,
                  frames=True,
                  forms=True,
                  annoying_tags=True,
                  remove_unknown_tags=True,
                  safe_attrs_only=True,
                  safe_attrs=frozenset(['src','color', 'href', 'title', 'class', 'name', 'id']),
                  remove_tags=('span', 'font', 'div')
                  )

    return cleaner.clean_html(dirty_html)
```

```python
onlyfiles = [f for f in listdir(path_to_html_files) if isfile(join(path_to_html_files, f))]
```

```python
for file in tqdm(onlyfiles):
    dirty_html = open(path_to_html_files+file, 'r', encoding="utf8")
    dirty_html_text = dirty_html.read()
    dirty_html.close()
    
    f = open(path_to_clean_html_files+"clean_"+file, "x", encoding="utf8")
    f.write(sanitize(dirty_html_text))
    f.close()
```
